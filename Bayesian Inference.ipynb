Bayesian inference is a method of inferential Statistics in which Bayes' Theorem is used to update the probability of a hypothesis or model as the information becomes available. Let's talk about the conditional probablity and how it is related with Bayes' theorem.

### <span style="color:#BB1C75">Conditional Probability</span>

The probability of occurrence of any event when another event in relation with that event has already occurred is known as conditional probability.

For example, probability of A, given B has occured, denoted by P(A$\mid$B). From Set Theory, we can define P(A$\mid$B) as the probability of both A and B are both occured devided by the probability of B occured.

$\hspace{8cm} P(A|B) = \Huge \frac{P(A\cap B)}{P(B)}$

Similarly $\small P(B|A) = \large \frac{P(B\cap A)}{P(A)}$ and we can say that P(A$\mid$B) and P(B$\mid$A) are same. Now we can write, P(A|B) P(B) = P(B|A) P(A)

So we can conclude that $\;\;\;P(A|B) = \Large \frac{P(B|A)P(A)}{P(B)}$ , $\hspace{0.1cm}$ where P(A|B) = Posterior probability, P(B|A) = Likelihood, P(A) = Prior probability, P(B) = Evidence or Marginal Likelihood.

This is known as $\color{#06AB80}{\text{Bayes' Theorem}}$. $\;$In general, $\hspace{0.1cm} Posterior$ =  <span style="font-size:20px;">$\frac{Likelihood\;\times\;Prior}{Evidence}$</span>

#####

### <span style="color:#BB1C75">Some examples</span>

##### <span style="color:#0F65B1">What is the probability of drawing a red-suit card from a pack of 52 cards ?</span>

$P(red\;card | T)$ = <span style="font-size:18px;">$\frac{26}{52}$ = $\frac{1}{2}$</span> $\;$, where T $\equiv$ Total number of card

##### <span style="color:#0F65B1">What is the probability of drawing two red-suit card from a pack of 52 cards ?</span>

The probability of occuring one red card is $R_0$ = <span style="font-size:18px;">$\frac{26}{52} = \frac{1}{2}$</span> and then another red card is $R_1$ = <span style="font-size:18px;">$\frac{26-1}{52-1} = \frac{25}{51}$</span>. So, the total probability of occuring both can be written as, $R_0 \times R_1$ = <span style="font-size:18px;">$\frac{25}{102}$</span>

Another way to calculate,

$P(R_0 , R_1|T) = P(R_1|R_0,T)\;P(R_0|T)$ = <span style="font-size:18px;">$\frac{25}{51} \times \frac{1}{2} = \frac{25}{102}$</span>

### <span style="color:#BB1C75">Parameter Estimation</span>

Parameter estimation is a process used in statistics and machine learning to estimate the values of parameters of a mathematical model that better explain the data.

$\hspace{8cm}$ <span style="font-size:18px;">$P(\theta|d,M)$</span> = <span style="font-size:25px;">$\frac{\mathcal{L}(d|\theta,M)\;\ \Pi(\theta|M)}{P(d|M)}$</span>


where d = observed data, $\theta$ = Parameters and M = Model hypothesis

So we need to find all the probabilistic quantity such as Likelihood, Prior and evidence.

##### <span style="color:#4DA30B">Step-1 : Define Likelihood ($\mathcal{L}$)</span>

It is better to choose your likelihood as a guassian distribution because -

<ul>
    <li>The Gaussian distribution has good mathematical properties, such as symmetry and a well-defined mean and variance. These properties make it easier to handel.</li>
    <li>Many natural and measurement processes result in data that can be approximated by a normal distribution.</li>
</ul>

The likelihood of the data point $y_i,\; t_i,$ given a value for the coefficients we will define to be Gaussian distributed as such :

$ P(y_i, t_i|\theta,\;M)$ = <span style="font-size:22px;">$\frac{1}{\sqrt{2\pi \sigma^2}}$</span> <span style="font-size:18px;">$\exp{(\frac{-(y_i-M)^2}{2 \sigma^2})}$</span>

Assume that all data points are independent. So, the total probability, $P(\{y, t\}|\theta,\;M) = \prod_{i=1}^nP(y_i, t_i|\theta,\;M)$

When solving problems on a computer, it is often convenient to work with the log-likelihood because it provides many advantages in terms of mathematical simplicity, numerical stability, and also it simplifies the multiplication of probabilities (especially with large datasets) into a sum of logarithms which provide computational efficiency.

For the normal distribution, the log-likelihood for n data points is

$$\log(P(\{y_i, t_i\}|\theta,\;M)) = - \frac{1}{2} \left[\sum_{i=1}^{n} \left(\frac{-(y_i-M)^2}{\sigma^2}\right)+n\log(2\pi\sigma^2)\right]$$

But we need Likelihood not Log-Likelihood ! 

So, just take an exponential of those values. Now, mean-centering is sometimes used in the context of converting Log-Likelihood to Likelihood to prevent extremely small or large Likelihood values (Exponentiating large negative numbers results in values close to zero) and provide us more manageable exponentiated data.

In order to do mean-centering, we just need to substract the mean of the samples from each and every sample. After that, we can take the exponential and define our Likelihood. 

We normalize the Likelihood, divide each Likelihood value by the sum of the Likelihoods, to interprete it as a probability distribution (Value = 0 to 1) and use in further statistical analysis. 

######

##### <span style="color:#4DA30B">Step-2 : Define Prior ($\Pi$)</span>

On the basis of our knowledge about the model, we define the normalized prior distribution. There are many possible ways to consider it such as Beta distribution, Uniform distribution and Gaussian distribution.

######

##### <span style="color:#4DA30B">Step-3 : Calculate Evidence</span>

This is the probability of the data when the model is true. It is computed by integrating (or summing) the likelihood times the prior over all possible values of parameters.

For continuous parameters, <span style="font-size:16px">$P(d|M) = \int \mathcal{L}(d|\theta,M)\;\ \Pi(\theta|M)\;d\theta$</span> $\;$ 

For discrete parameters, <span style="font-size:16px">$P(d|M) = \sum_{\theta}\;\mathcal{L}(d|\theta,M)\;\ \Pi(\theta|M)$</span>

This integral can be extremely difficult to compute analytically, especially for high-dimensional parameter spaces.

######

##### <span style="color:#4DA30B">Step-4 : Calculate Posterior</span>

Posterior is a understanding of how the data is distributed given that we witnessed the data, and that we had a prior about it. Multiply the Likelihood with the Prior and divide it by Evidence, we will get the normalized posterior distribution. 

######

##### <span style="color:#4DA30B">Step-5 : Apply Metropolis-Hastings algorithm</span>

The Metropolis-Hastings algorithm is a popular MCMC (Markov Chain Monte Carlo) method. It is used to generate a Markov chain of sample values from a target probability distribution.

Here is a detailed description of the Metropolis-Hastings algorithm,

<span style="color:#BDA249">Useful Components -</span>
<ul>
    <li><i>Target Distribution</i>, P(x) : The distribution from which we want to sample. It is often taken in the form of an unnormalized probability density function.</li>
    <li><i>Proposal Distribution</i>, Q($x^{\prime}$|x) : A conditional probability distribution used to propose a new state $x^{\prime}$ given the current state. It's a distribution used to generate candidate samples that are then either accepted or rejected based on the acceptance condition. The proposal distribution can influence the efficiency of the algorithm.</li>
</ul>

<span style="color:#BDA249">Algorithm Steps -</span>

<ol>
    <li> <i>Initialization</i> : Start at an initial state $x_0$ and t=0</li>
    <li> <i>Iteration</i> : </li>
    <ul>
        <li>Sample a candidate state $x^{\prime}$ from the proposal distribution Q($x^{\prime}$|$x_t$).</li>
        <li>Calculate the acceptance ratio ($\alpha$) : <br> $\hspace{6cm}$ $\large\alpha$ = $\large\min$ <span style="font-size:22px">$\left(1, \frac{P(x^\prime)}{P(x_t)}\frac{Q(x_t|x^{\prime})}{Q(x^{\prime}|x_t)}\right)$</span> </li>
        <li>Make acceptance decision : </li>
        <ul type ="A">
            <li>Generate a random number $\large u$ from the uniform distribution $U(0,1)$.</li>
            <li>If $\large u \leq \large\alpha$, accept the new state $x^\prime$ and set $x_t=x^\prime$.</li>
            <li>Otherwise, reject $x^\prime$ and set $x_{t+1}=x_t$</li>
        </ul>
    </ul>
    <br><li><i>Update</i> : Repeat this process until the desired number of samples is collected.</li>
</ol>

So, the Metropolis-Hastings algorithm provides a flexible framework for sampling from complex distributions, and it is widely used in various fields such as Bayesian Statistics, Machine Learning, and Physics.

**Note :** 

<ul>
    <li>MCMC method do not consider the Evidence or Marginal Likelihood because it is a constant factor (does not depend on parameter) for the posterior distribution and also it reduces the computational burden.</li>
    <li>If the proposal distribution Q($x^{\prime}$|x) is symmetric, i.e. Q($x^{\prime}$|x) = Q(x|$x^{\prime}$), the proposal terms cancel out in the acceptance ratio then we can neglect the proposal distribution.</li>
</ul>

## <span style="color:#BB1C75">Python for one-dimensional Posterior distribution</span>

######

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

#####

Let's take a model of sinusoidal function, <span style="font-size:15px">$y_i = \sin(\omega t_i)$</span>. Here we have one parameter, i.e. $\large \omega$

```python
def model(omega, t):
	return np.sin(omega * t)
```

#####

Choose a particular $\large \omega$ (Let's say 5.5) for different time and mix Gaussian noise with it to generate the data

```python
time = np.linspace(0, 12, 1000)

sigma = 0.5
omega_true = 5.5

y_signal = model(omega = omega_true, t = time)
y_obs = y_signal + 5*np.random.normal(0, sigma, len(time))

plt.plot(time, y_obs, label='Data')
plt.plot(time, y_signal, label='Model')
plt.legend()
```

#

Make Log-Likelihood function

```python
def log_likelihood(y_obs, time, omega, sigma = 0.1):
	y_A = model(omega,time)
	L = -0.5 * ((y_obs - y_A)**2/sigma**2 + np.log(2 * np.pi * sigma**2))
	return np.sum(L)
```

######

Run the Log-Likelihood function for different $\large \omega$ value

```python
omega_array = np.linspace(0, 10, len(time))

LL_array = []
for omega in omega_array :
	LL_array.append(log_likelihood(y_obs, time, omega))

plt.plot(omega_array,LL_array)
plt.xlabel('$\omega$')
plt.ylabel('Log-Likelihood')
```

#

We see that there exist a peak for the maximum Log-Likelihood value associated with a $\large \omega$  value. We only consider a small interval around that $\large \omega$ value.

```python
Max_omega = omega_array[np.argmax(LL_array)]

omega_array = np.linspace(Max_omega-2e-2, Max_omega+2e-2, len(time))

LL_array = []
for omega in omega_array :
	LL_array.append(log_likelihood(y_obs, time, omega))

Max_omega = omega_array[np.argmax(LL_array)]
print('\n The omega value which corresponds to that peak : ', Max_omega, '\n')

plt.plot(omega_array, LL_array)
plt.axvline(omega_true, c='red', label='Injected')
plt.axvline(Max_omega, c='green', label='Max_LL')
plt.xlabel('$\omega$')
plt.ylabel('Log-Likelihood')
plt.legend()
```

#

Now, perform mean-centering (which I have discussed before) and convert the log-likelihood into likelihood.

```python
Mean_Center_LL = LL_array - np.mean(LL_array)

Likelihood = np.exp(Mean_Center_LL)
Likelihood_norm = Likelihood / np.sum(Likelihood)

plt.plot(omega_array, Likelihood_norm)
plt.xlabel('$\omega$')
plt.ylabel('Normalized Likelihood')
```

#

Here I set a Gaussian distribution ($\mu$=5.49, $\sigma$=0.001) for prior.

```python
prior = stats.norm.pdf(omega_array, 5.49, 0.001)
prior_norm = prior/np.sum(prior)

# for uniform distribution, prior = stats.uniform.pdf(omega_array) + 1
```

#####

Since $\large \omega$ is a discrete set, to calculate the marginal likelihood we should use summation rule.

```python
Evidence = np.sum(prior_norm * Likelihood_norm)
```

#####

Posterior is defined by multiplying the likelihood and the prior.

```python
posterior_unnorm = prior_norm * Likelihood_norm
posterior_norm = posterior_unnorm/Evidence
```

######

Let's plot all of three in one figure.

```python
plt.plot(omega_array, prior_norm, color='navy', label='Prior')
plt.plot(omega_array,Likelihood_norm, color='red', label='Likelihood')
plt.plot(omega_array, posterior_norm, color='green', label='Posterior')
plt.xlabel('$\omega$')

plt.legend()
```

####

<b>Note :</b> If the prior distribution is uniform then the posterior distribution will same as the likelihood.

######

### <span style="color:#BB1C75">Find the Markov Chain !</span>

To apply the sampling technique, it is covenient to use the logarithm of the probability distributions. Also we need to modify the algorithm steps for logarithmic version.

I already mention before that how the Log-Likelihood looks like. But this time, we don't need the array of $\large\omega$, only consider it for one value at a time.

```python
def Log_Likelihood(y, time, omega, sigma = 0.1) :
	y_A = model(omega,time)
	LL = np.sum(-0.5 * (((y - y_A)**2)/sigma**2 + np.log(2 * np.pi * sigma**2)))
	return LL
```

#####

Now we have to consider the logarithm of the prior distribution. Here I choose the Prior as a Gaussian distribution of the mean value corresponds to maximum Log-Likelihood. Otherwise, we can choose the Prior based on our knowledge about the model parameter.

```python
def Log_Prior(omega, mean = Max_omega, sigma = 0.001) :
	x = stats.norm.pdf(omega, mean, sigma)
	return np.log(x)
```

#####

Similarly, for Log-Posterior, we can write it as the sum of Log-Likelihood and Log-Prior.

```python
def Log_Posterior(y, time, omega) :
    return Log_Prior(omega) + Log_Likelihood(y, time, omega)
```

#####

Now, we will make a function which gives the Markov chain. 

Here, I take the proposal distribution as a symmetric distribution centered around the current state, such as $x^\prime = x_t + \epsilon$, where $\epsilon$ is drawn from a symmetric distribution (normal distribution with zero mean). So, Q($x^{\prime}$|x) = Q(x|$x^{\prime}$), we can neglect their ratio.

We must consider the Log-Alpha which can be written as the sum of Log-Posteriors of new state and proposed state. Then we put the acceptance condition between the Log-Alpha and the log value of a random number from an uniform distribution $U(0,1)$.

**Note :** 

<ul>
    <li>It is convenient to set the sample step_size (basically it is the standard deviation) of proposal distribution as same as the Prior's. Otherwise, $\log{(0)}$ is an undefined quantity, it will give an error.</li>
    <li>If the Prior is taken as an uniform distribution, then we can consider the Likelihood as the target distribution.</li>
</ul>

```python
def Metropolis_Hastings(Log_Posterior, y, time, omega_in, step_size, iteration) :
	omega_current = [omega_in]

	for _ in range(iteration):
		omega_proposed = omega_current[-1] + np.random.normal(0, step_size)
		Log_Alpha = Log_Posterior(y, time, omega_proposed) - Log_Posterior(y, time, omega_current[-1])
		if Log_Alpha > np.log(np.random.rand()) :
			omega_current.append(omega_proposed)
	
	return omega_current
```

#####

Set the functional values and plot the chain !

```python
y = y_obs
omega_in = Max_omega-2e-2
step_size = 0.001
iteration = 10000

samples = Metropolis_Hastings(Log_Posterior, y, time, omega_in, step_size, iteration)

plt.plot(samples, '--+', markeredgecolor='darkorchid', color='seagreen', lw=0.7)
plt.ylabel('$\omega$')
plt.xticks([])
```

####

```python
plt.boxplot(samples)
```

### <span style="color:#BB1C75">Get rid of Burn-In phase</span>

The burn-in phase refers to the initial period during which the Markov chain might not have reached its stationary distribution. Samples taken during this period can be biased and are typically discarded to ensure the quality of the final samples. Generally the outliers can be considered as the burn-in phases.

We can apply the quantile rules to remove most of those outliers.

<span style="font-size:15px">IQR = Q3 - Q1</span>

<span style="font-size:15px">Lower Fench = Q1 - (1.5 * IQR)</span>

<span style="font-size:15px">Higher Fench = Q3 + (1.5 * IQR)</span>

```python
Q1, Q3 = np.percentile(samples, [25,75])

IQR = Q3 - Q1
Lower_fench = Q1 - (1.5 * IQR)
Higher_fench = Q3 + (1.5 * IQR)

print(Q1, Q3, Lower_fench, Higher_fench)
```

######

Now, we need to put the boundary condition.

```python
samples.sort()
samples = samples[np.where(samples>Lower_fench)[0][0]:np.where(samples<Higher_fench)[0][-1]]

plt.boxplot(samples, showmeans=True)
```

Now, It looks more convenient.

## <span style="color:#BB1C75">What about two-dimensional posterior ?</span>

######

<b>Model :</b> <span style="font-size:16px">$y_i = mx_{i} + c$</span> ; Now we have two parameter $m$ and $c$

```python
def model(m, c, x):
    return m*x + c
```

#####

Make the noisy data --

```python
m_true = 3
c_true = 8
sigma = 0.5

x = np.linspace(0,20,1000)

y_signal = model(m = m_true, c = c_true, x = x)
y_obs = y_signal + 10*np.random.normal(0, sigma, len(x))

plt.plot(x, y_obs, label='Data')
plt.plot(x, y_signal, label='Model')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```

###

We can define the Log-Likelihood function in the same way $\;$ ---

```python
def Log_Likelihood(y, x, m, c, sigma=0.1) :
	y_A = model(m, c, x)
	L = -0.5 * (((y - y_A)**2)/sigma**2 + np.log(2 * np.pi * sigma**2))
	return np.sum(L)

Array = np.linspace(0, 10, len(x))

LL_array = np.zeros((len(x),len(x)))

for i in range(len(x)) :
    for j in range(len(x)) :
        LL_array[i,j] = Log_Likelihood(y_obs, x, m=Array[i], c=Array[j], sigma=0.1)
```

#####

Estimate the value of $m$ and $c$ for maximum Log-Likelihood $\;$ ---

```python
pos = np.argmax(LL_array)
row = pos//len(x)
col = pos % len(x)

Max_m = Array[row]
Max_c = Array[col]

print(Max_m, Max_c)
```

######

Visualization $\;$ ---

```python
fig, ax = plt.subplots(1,1)

f = ax.pcolor(Array, Array, LL_array, cmap='winter')
ax.plot(c_true, m_true, '.', markersize=8, c='red', label='Injected')
ax.plot(Max_c, Max_m, '+', markersize=8, c='k', label='Max_LL')
ax.set_xlabel('$c$')
ax.set_ylabel('$m$')
leg = ax.legend(loc='upper left')
cbar = fig.colorbar(f, label='Log_Likelihood')
```

#####

We can define Gaussian Prior in the following way  $\;$ ---

```python
def Log_Prior(m, c, mean_m=Max_m, mean_c=Max_c, sigma=0.01) :
    x = stats.norm.pdf(m, mean_m, sigma)
    y = stats.norm.pdf(c, mean_c, sigma)
    return np.log(x)+np.log(y)
```

#####

But here I choose the Prior as an uniform distribution. We will consider the Posterior as the Likelihood function.

```python
def Log_Posterior(y, x, m, c) :
    return Log_Likelihood(y, x, m, c)
```

#####

Apply MCMC $\;$ ---

```python
def Metropolis_Hastings(Log_Posterior, y, x, m_in, c_in, step_size, iteration) :
    m_current = [m_in]
    c_current = [c_in]

    for _ in range(iteration):
        m_proposed = m_current[-1] + np.random.normal(0, step_size)
        c_proposed = c_current[-1] + np.random.normal(0, step_size)
        Log_Alpha = Log_Posterior(y, x, m_proposed, c_proposed) - Log_Posterior(y, x, m_current[-1], c_current[-1])
        if Log_Alpha > np.log(np.random.rand()) :
            m_current.append(m_proposed)
            c_current.append(c_proposed)
	
    return m_current, c_current
```

```python
y = y_obs
m_in = Max_m-2e-1
c_in = Max_c-2e-1
step_size = 0.01
iteration = 100000

m_samples, c_samples = Metropolis_Hastings(Log_Posterior, y, x, m_in, c_in, step_size, iteration)
```

#####

Now let's see the Markov Chain for each parameters $\;$ ---

```python
fig, ax = plt.subplots(1,2, figsize=[10,4])

ax[0].plot(m_samples, '--*', markeredgecolor='darkseagreen', color='sienna', lw=0.7)
ax[0].set_ylabel('$m$')
ax[0].set_xticks([])

ax[1].plot(c_samples, '--*', markeredgecolor='darkseagreen', color='sienna', lw=0.7)
ax[1].set_ylabel('$c$')
ax[1].set_xticks([])
```
